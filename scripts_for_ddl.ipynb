```python
table_name = 'dds.stock_dividends_history'
years = [2000 + i for i in range(26)]
full_sql = ""

for year in years:
    full_sql += f"""
       CREATE TABLE {table_name}_{year} PARTITION OF {table_name} FOR VALUES FROM ('{year}-01-01') TO ('{year}-12-31');"""
    
print(full_sql)
```
